Define $W(t)$ as a standerd brownian motion that is adaptable to the filtration.
By means of a Monte Carlo simulation, show that 
$$ E[W(t)|F(t_{0})] = W(t_{0})$$
$$ E[W(t)|F(s)] = W(s), \text{with s < t}$$

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
t = 10
s = 5
NoOfPaths = 10000
NoOfSteps = 10
```

```python
def martingaleA():
    W_t = np.random.normal(0.0, pow(t,0.5), [NoOfPaths,1]) # Generate 10000 (NoOfPaths) paths with a granularity of 1
    E_W_t = np.mean(W_t) # Take the expected value, should be 0
    print("mean value equals to: {}".format(round(E_W_t, 4)))

martingaleA() 
# mean value equals to: -0.0343; close to 0 as expected (No seed set, will generate differnt values)
```

```python
# Creating a nested Monte Carlo Simulation E(W(t)|F(s))=W(s)
def martingaleB():
    Z = np.random.normal(0, 1, [NoOfPaths,NoOfSteps]) # Simulating brownian motion with a granularity of 10 (NoOfSteps)
    W = np.zeros([NoOfPaths, NoOfSteps+1]) # making a matrix of zeros since values unitl time s doesn't matter
    # Simulation from time [0,s]
    dt1 = s/float(NoOfSteps) # time steps between [0,s]
    for i in range(0,NoOfSteps):

        Z[:,i] = (Z[:,i] - np.mean(Z[:,i]))/np.std(Z[:,i]) # Standerdize Z to be N(0,1)
        W[:,i+1] = W[:,i] + pow(dt1,0.5)*Z[:,i]

    W_s = W[:,-1] #last column of W is the value W_s
    
    # For every path W(s) we make sub-simulations until time t
    # time steps betweem [s,t]
    dt2 = (t-s)/float(NoOfSteps);
    W_t = np.zeros([NoOfPaths, NoOfSteps+1])

    # To store the results
    E_W_t = np.zeros([NoOfPaths]) 
    Error = []
    for i in range(0,NoOfPaths):
        W_t[:,0] = W_s[i]; # first value of W_t is each endpoint at generated at W_s
        Z = np.random.normal(0,1,[NoOfPaths,NoOfSteps])
        for j in range(0,NoOfSteps):
            Z[:,j] = (Z[:,j] - np.mean(Z[:,j]))/np.std(Z[:,j]); # Standerdize 
            W_t[:,j+1] = W_t[:,j] + pow(dt2,0.5)*Z[:,j]; # simulate from s to t

        E_W_t[i] = np.mean(W_t[:,-1])
        Error.append(E_W_t[i]-W_s[i])
    
        if i==0:
            plt.plot(np.linspace(0,s,NoOfSteps+1),W[0,:])
            for j in range(0,NoOfPaths):
                plt.plot(np.linspace(s,t,NoOfSteps+1),W_t[j,:])
            plt.xlabel("time")
            plt.ylabel("W(t)")
            plt.grid()
        
    print(Error)
    error = np.max(np.abs(E_W_t-W_s))
    print("The error is equal to: %.18f"%(error))
    
martingaleB()
```
